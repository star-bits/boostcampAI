````python
# numpy

# dot - 일반적인 matrix multiplication

a1 = np.arange(6).reshape(2, 3)
a2 = np.arange(6).reshape(3, 2)

a = np.dot(a1, a2)
a
````

````python
# @ - 일반적인 matrix multiplication

a = a1 @ a2
a
````

````python
# inner - dimension 똑같아야

a = np.inner(a1, a2.T)
a
````

````python
# torch

# mm, matmul - 일반적인 matrix multiplication

t1 = torch.tensor(a1)
t2 = torch.tensor(a2)

t = torch.mm(t1, t2)
t
````

````python
# dot - 1D tensor일 경우에만 

t3 = torch.arange(3).reshape(3,)
t4 = torch.arange(3).reshape(3,)

t = torch.dot(t3, t4)
t
````

````python
# inner - dimension 똑같아야

t = torch.inner(t1, t2.T)
t
````

# Autograd

````python
# Autograd

w = torch.tensor(2.0, requires_grad=True)
y = w**2
z = 10*y + 2
z.backward()
w.grad
````

````python
# cost function이 없다고 어색한 게 아님.
# DL model은 미분머신이고, 어차피 미분값 찾는건 같은데
# 마지막에 cost function을 씌워줌으로서 cost에 대한 미분을 하게 바뀌는 것 뿐.
# label은 cost function을 계산하는 데 도와주는 것 뿐임.
# dL/dw 또는 dy/dw의 식을 구하고, 거기다 w를 대입해 기울기를 구하고,
# 그 기울기에다 learning rate를 곱해줘서 업데이트 하면 그게 GD. 
````

````python
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)
Q = 3*a**3 - b**2
# Q' = 9*a**2 - 2*b 
# 편미분인거 감안
external_grad = torch.tensor([1., 1.]) # ?
Q.backward(gradient=external_grad)
````
