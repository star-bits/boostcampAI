# Momentum



$$v_{dW} = \beta v_{dW} + dW$$
$$W = W - \alpha v_{dW} $$

v_{dW}: momentum
<br>
\beta: momentum hyperparameter
<br>
\alpha: learning rate

# RMSprop



$$s_{dW} = \beta_{2} s_{dW} + (1-\beta_{2}) dW^2$$
$$W = W - \alpha \frac{dW}{\sqrt{s_{dW}+\varepsilon }} $$

\beta_{2}: RMSprop hyperparameter

# Adam



$$v_{dW} = \beta_{1} v_{dW} + (1-\beta_{1}) dW$$
$$s_{dW} = \beta_{2} s_{dW} + (1-\beta_{2}) dW^2$$

$$v_{dW}^{corrected} = \frac{v_{dW}}{(1-\beta_{1}^{t})}$$
$$s_{dW}^{corrected} = \frac{s_{dW}}{(1-\beta_{2}^{t})}$$

$$W = W - \alpha \frac{v_{dW}^{corrected}}{\sqrt{s_{dW}^{corrected} + \varepsilon}} = W - \alpha \frac{v_{dW}}{s_{dW}} \frac{\sqrt{1-\beta_{2}^{t}}}{1-\beta_{1}^{t}}$$

Momentum + Adaptive learning rate
