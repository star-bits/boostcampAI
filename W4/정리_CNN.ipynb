# VGGNet: 3x3

### VGGNet uses kernels with size 3x3, as opposed to AlexNet's 11x11 kernel

3x3 twice has effectively the same 'receptive field' as 5x5 once, while having smaller number of parameters.

# ResNet: skip-connection

### More layers, worse performance

Concern for models with more layers is that it might cause overfitting. But that is not the issue here. The problem here is that deeper model is performing worse than shallower model, in *both* training and in test.


Skip connection solves the issue. (Skip connection on ResNet *adds* an indentity map.)

# DenseNet: skip-connection - concat

Skip connection on DenseNet *concatenates* an indentiy map. But then the number of channels increase geometrically! -> Use 1x1 Conv.
