```python
import matplotlib.pyplot as plt
```

```python
# figure에 subplot을 최소 1개 이상 추가하고, 이 figure를 출력하는 것

fig = plt.figure(figsize=(3, 2)) # figsize inches
ax = fig.add_subplot()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1) # 세로를 2로 나누고 가로를 1로 나눴을 때 1번째
ax2 = fig.add_subplot(2, 1, 2)
plt.show()
```

```python
# Pyplot API vs OO API

# Pyplot API

import numpy as np

x1 = np.array([1, 2, 3])
x2 = np.array([3, 2, 1])


fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
plt.plot(x1) # 가장 최근의 subplot에 plot함

ax2 = fig.add_subplot(2, 1, 2)
plt.plot(x2)

plt.show()
```

```python
# OO API

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x1)
ax2.plot(x2)

plt.show()
```

```python
# color

fig = plt.figure()
ax = fig.add_subplot()

ax.plot([1, 1, 1], color='r')
ax.plot([1, 2, 3], color='forestgreen')
ax.plot([3, 3, 3], color='#000000')

plt.show()
```

```python
# ax.set_title, ax.legend, fig.suptitle

fig = plt.figure()
ax = fig.add_subplot(111) 

ax.plot([1, 1, 1], label='1') 
ax.plot([2, 2, 2], label='2') 
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')

ax.legend()

fig.suptitle('Title') # super title

plt.show()
```

```python
print(ax.get_title())
```

```python
# ax.set_xticks

fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot([1, 1, 1], label='1') 
ax.plot([2, 2, 2], label='2') 
ax.plot([3, 3, 3], label='3')

ax.set_xticks([0, 1, 2, 3])

plt.show()
```

```python
# ax.set_xticklabels

fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot([1, 1, 1], label='1') 
ax.plot([2, 2, 2], label='2') 
ax.plot([3, 3, 3], label='3')

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(['zero', 'one', 'two', 'three'])

plt.show()
```

```python
# ax.text

fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot([1, 1, 1]) 
ax.plot([2, 2, 2]) 
ax.plot([3, 3, 3])

ax.text(x=1, y=2, s='This is ax.text()')

plt.show()
```

```python
# ax.annotate

fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot([1, 1, 1]) 
ax.plot([2, 2, 2]) 
ax.plot([3, 3, 3])

ax.annotate(text='This is ax.annotate()', xy=(1, 2))

plt.show()
```
