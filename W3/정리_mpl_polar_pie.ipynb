# Polar

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, polar=True)

ax.set_rmax(2)
ax.set_rmin(1)
ax.set_rticks([1, 1.5, 2])

ax.set_rlabel_position(-90)

ax.set_thetamin(45)
ax.set_thetamax(315)

plt.show()
```

```python
N = 100
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, polar=True)
ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.5)
plt.show()
```

```python
N = 6
r = np.random.rand(N)
theta = np.linspace(0, 2*np.pi, N, endpoint=False)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, polar=True)
ax.bar(theta, r, width=0.5, alpha=0.5)
plt.show()
```

```python
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
```

```python
# 끝 점을 포함하기 위해 values, theta 수정
values = pokemon.iloc[0][stats].to_list()
values.append(values[0])
theta = np.linspace(0, 2*np.pi, 6, endpoint=False)
theta = theta.tolist() + [theta[0]]

print(values)
print(theta)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, polar=True)

ax.plot(theta, values)
ax.fill(theta, values, alpha=0.5)

ax.set_thetagrids([n*60 for n in range(6)], stats)
ax.set_theta_offset(np.pi/2)

plt.show()
```

```python
fig = plt.figure(figsize=(14, 4))

for idx in range(3):
    ax = fig.add_subplot(1, 3, idx+1, polar=True)
    
    values = pokemon.iloc[idx][stats].to_list()
    values.append(values[0])
    
    ax.plot(theta, values, color='forestgreen')
    ax.fill(theta, values, color='forestgreen', alpha=0.3)
    
    ax.set_rmax(100)
    ax.set_thetagrids([n*60 for n in range(6)], stats)
    ax.set_theta_offset(np.pi/2)

plt.show()
```

# Pie

```python
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
labels = ['A', 'B', 'C', 'D']
data = np.array([60, 90, 45, 165])
```

```python
explode = [0, 0, 0.2, 0]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.pie(data, labels=labels, 
       explode=explode, 
       startangle=90, 
       autopct='%1.1f%%',
       pctdistance=0.75,
       textprops={'color':'blue'},
       labeldistance=1.25,
       rotatelabels=0,
       counterclock=False,
       radius=0.75)

center_circle = plt.Circle((0,0), 0.3, fc='white')
ax.add_artist(center_circle)

plt.show()
```
