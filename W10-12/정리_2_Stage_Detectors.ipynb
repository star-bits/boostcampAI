# R-CNN

1. selective search - 2000 ROIs
2. warp images - same FC layer output dim
3. 4096-dim feature vector per region
4. classification with SVM
5. bounding box regression (refinement)

### shortcomings:

- 2000 CNN operations
- image warping
- separate training for CNN, SVM classifier, bounding box regressor
- not an end-to-end learning

# SPPNet

- R-CNN: image -> ROI (selective search) -> conv -> fc -> output
- SPPNet: image -> conv -> ROI projection -> spatial pyramid pooling -> fc -> output

### ROI projection

proportional projection of selective search result (of input image) on the feature map



### SPP

- input: 64x64x256
- three spatial pyramid layers grid size: 4x4, 2x2, 1x1
- pooling output: 16x256, 4x256, 1x256
- final output: (16+4+1)x256 dim vector



### shortcomings solved:

- 2000 CNN operations
- image warping

# Fast R-CNN

### ROI pooling instead of spatial pyramid pooling

difference: one spatial pyramid layer grid size only - 7x7

### Multi-task loss

- classification + bounding box regression
- classification loss function: cross entropy
- bounding box regression loss function: smooth L1

### shortcoming solved:

- separate training for CNN, SVM classifier, bounding box regressor

### still using selective search

- selective search is not trainable
- uses CPU

# Faster R-CNN

- Fast R-CNN ROI: image -> selective search -> ROI projection on feature map
- Faseter R-CNN ROI: image -> feature map -> Region Proposal Network

### Region Proposal Network

image -> feature map -> cells -> anchor boxes per cell

RPN determines whether each anchor box contains an object and if it does, performs bounding box refinement.

RPN consists of cls layer (classification head) and reg layer (regression head)



9 anchor boxes per cell (in this case, each pixel is a single cell)
