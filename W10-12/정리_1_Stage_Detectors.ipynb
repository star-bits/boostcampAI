# YOLOv1

### Loss

Localization loss + Confidence loss + Classification loss



### Localization loss

각 grid cell의 각 box 별로

obj를 포함하고 있을 때

중심점 (x, y)의 위치 계산

+

각 grid cell의 각 box 별로

obj를 포함하고 있을 때

w, h의 길이 계산

(lambda coord는 hyperparameter)

### Confidence loss

각 grid cell의 각 box 별로

obj를 포함하고 있을 때

confidence score를 계산 

+

각 grid cell의 각 box 별로

obj를 포함하고 있지 않을 때

confidence score를 계산 

(lambda noojb는 hyperparameter)

### Classification loss

각 grid cell 별로

obj를 포함하고 있을 때 

classification MSE

# SSD

- 6개의 서로 다른 scale의 feature map 사용 (multi-scale feature maps)
  - FPN처럼 early stage에서는 작은 물체, late stage에서는 큰 물체 탐지
- fc layer 대신 conv layer 사용해서 속도 향상
- anchor box 사용

# YOLOv2

- no fc layer
- anchor box
- backbone: GoogLeNet -> Darknet-19

# RetinaNet

- 1 stage detector의 문제: Region Proposal Network가 없어서 그리드별로 bounding box를 무조건 예측해야 함
  - 백그라운드를 포함할 확률이 높아짐.
  - class imbalance
  - focal loss로 해결

- multi-scale feature maps
