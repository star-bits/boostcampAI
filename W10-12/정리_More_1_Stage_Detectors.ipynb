# One-stage detector vs Two-stage detector



Backbone
- GPU platform: VGG, ResNet, DenseNet, ...
- CPU platform: SqueezeNet, MobileNet, SutffleNet, ...

Neck
- Additional blocks: SPP, ASPP, ...
- Path-aggression blocks: FPN, PAN, NAS-FPN, BiFPN, ...

Head
- Dense prediction (one-stage): YOLO, SSD, RetinaNet, CornerNet, FCOS, ...
- Sparse prediction (two-stage): Faster R-CNN, R-FCN, Mask R-CNN, ...
